```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import RandomOverSampler

from tqdm.auto import tqdm
tqdm.pandas()
```

```python
train_1 = pd.read_json('Data/train.json')
```

```python
train_1
```

```python
def GetArticles(x):
    articles = []
    max_len = 5
    
    for i in x:
        source = open('Data/train_articles/{}.txt'.format(i), 'r', encoding = 'utf-8').read(-1)
        articles.append(source)
         
    return articles
```

```python
temp = train_1['related_articles'][:].progress_apply(GetArticles)
```

```python
train_2 = train_1.join(pd.DataFrame(temp.values.tolist()))
```

```python
train_2.to_csv("Data/ArticlesData.csv", header = True)
```

# Oversampling [UNUSED]

```python
train_3 = pd.read_csv('Data/ArticlesData.csv')
```

```python
train_3 = train_3.drop('Unnamed: 0', axis = 1)
train_3 = train_3.fillna(0)
```

```python
ros = RandomOverSampler(random_state = 4)
```

```python
X, y = ros.fit_resample(train_3.drop('label', axis = 1), train_3['label'])
```

```python
data = []
for i in range(len(X)):
    data.append(np.append(X[i], y[i]))
```

```python
columns = train_3.columns
```

```python
columns = columns.drop('label').append(pd.Index(['label']))
```

```python
train_4 = pd.DataFrame(data, columns = columns)
```

```python
train_4.to_csv('Data/ArticlesData1.csv')
```
